```python
cadastro01 = {'Matricula': 2000168933,
              'diaCadastro': 25,
              'mêsCadastro': 10,
              'turma': '2E'}
cadastro01
```

```python
cadastro01['turma'] = 'new value'
```

```python
cadastro01.clear()
```

```python
# metodo items(), retorna as chaves e valores de um dicionario

cadastro01.items()
```

```python
# metodo keys, retorna apenas as chaves do meu dicionario

cadastro01.keys()
```

```python
# metodo values, retona apenas os valores do meu dicionario

cadastro01.values()
```

```python
# usando for para percorrer um dicionario

for valores in cadastro01.values():
    print(valores)
```

```python
# for para keys

for chave in cadastro01.keys():
    print(chave)
```

```python
# for na chave:valor

for chave,valor in cadastro01.items():
    print(chave , valor)
```

```python
# lista dentro dos dicionarios

list = {'produtos': ['smartTv', 'Iphone','coffe','noteBook'],
        'preco': [2000,2500,120,5000]}

for produto,preco in list.items():
    print(produto,preco)
```
