```python
# Operadores Logicos 

# and só vai ser valido se as duas condicoes forém verdadeira

t1 = t2 = True
f1 = f2 = False


if t1 and f1: 
    print('expressao verdadeira')
else:
  print('expressao falsa')
```

```python
# o or ele aceita um ou outro

if f2 or f1: 
    print('expressao verdadeira')
else:
  print('expressao falsa')
```

```python
#  ele inverte a ordem se for true vira false

if not t1: 
    print('expressao verdadeira')
else:
  print('expressao falsa')
```
